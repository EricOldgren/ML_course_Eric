```python
## Task A ##

import numpy as np

def standardize(D):
    zero_mean_D = D - np.mean(D,axis = 0)
    unit_variance_D = zero_mean_D / np.std(zero_mean_D,axis = 0)
    print(np.std(unit_variance_D,axis=0))
    print(np.mean(unit_variance_D,axis=0))
    return unit_variance_D


D = np.random.randn(4,6)
print(D)
D_standard = standardize(D)
print(D_standard)
```

```python
## Task B ##

M1 = np.random.randn(2,1)
M2 = np.random.randn(2,4)

def pairwise_dist(M1,M2):
    N1 = np.size(M1,1)
    N2 = np.size(M2,1)
    distance_matrix = np.zeros([N1,N2])
    for i in range(N1):
        for j in range(N2):
            distance_matrix[i,j] = np.linalg.norm(M1[:,i]-M2[:,j])
    return distance_matrix

print(pairwise_dist(M1,M2))
```
